```python
ID = []
EFT_ERT = []
DELAY = []
EXCITEMENT = []
VIVID = []
IP100_VALUES = []
IP = []
IP1000_VALUES = []
CUES = []

count = 1
id_update = 1
for i in range(len(raw_data)):
#     if count > 3:
#         count = 1
#     id_update += 1
        
    if category[i] == 'EFT':
        delay = 0
        ip100 = 0.0
        ip1000 = 0.0 
        cue = []
        excited = 0.0
        vividness = 0.0
        eft = category[i]
        
        for j in range(1,6):
            if j == 1:
                delay = 30
                ip100 = IP30_100[i]
                ip1000 = IP30_1000[i]
                cue = MONTH1_CUE[i]
                excited = exciting_1[i]
                vividness = vivid_1[i]
            elif j == 2:
                delay = 90
                ip100 = IP90_100[i]
                ip1000 = IP90_1000[i]
                cue = MONTH3_CUE[i]
                excited = exciting_3[i]
                vividness = vivid_3[i]
            elif j == 3:
                delay = 365
                ip100 = IP365_100[i]
                ip1000 = IP365_1000[i]
                cue = YEAR_CUE[i]
                excited = exciting_yr[i]
                vividness = vivid_yr[i]
            elif j == 4:
                delay = 1825
                ip100 = IP1825_100[i]
                ip1000 = IP1825_1000[i]
                cue = YEAR5_CUE[i]
                excited = exciting_yr5[i]
                vividness = vivid_yr5[i]
            elif j == 5:
                delay = 9125
                ip100 = IP9125_100[i]
                ip1000 = IP9125_1000[i]
                cue = YEAR25_CUE[i]
                excited = exciting_yr25
                vividness = vivid_yr25

            ID.append(i+1)
            EFT_ERT.append(eft)
            DELAY.append(delay)
            IP100_VALUES.append(ip100)
            IP1000_VALUES.append(ip1000)
            IP.append('IP' + str(delay))
            CUES.append(cue)
            EXCITEMENT.append(excited)
            VIVID.append(vividness)
        
#         id_update += 1
#         count += 1
```

# STUDY 6 

```python
raw_data = pd.read_excel('STUDY6.xlsx')
raw_data = pd.DataFrame(raw_data)
raw_data
```

```python
SID = raw_data['SUBJECTID']
GROUP = raw_data['GROUP']

MONTH_G1_CUE = raw_data['G1CUE2']
MONTH_G2_CUE = raw_data['G2CUE2']
YEAR_G2_CUE = raw_data['G2CUE3']
YEAR_G1_CUE = raw_data['G1CUE3']

IP30 = raw_data['IP30']
IP180 = raw_data['IP180']
IP365 = raw_data['IP365']

exciting_G1_month = raw_data['G1RATING2_3']
exciting_G2_month = raw_data['G2RATING2_3']
exciting_G2_yr = raw_data['G2RATING3_3']
exciting_G1_yr = raw_data['G1RATING3_3']

vivid_G1_month = raw_data['G1RATING2_4']
vivid_G2_month = raw_data['G2RATING2_4']
vivid_G1_yr = raw_data['G1RATING3_4']
vivid_G2_yr = raw_data['G2RATING3_4']
```

```python
ID = []
EFT_ERT = []
DELAY = []
IP_VALUES = []
IP = []
CUES = []
EXCITEMENT = []
VIVID = []

for i in range(len(raw_data)):
    delay = 0
    ip = 0.0
    cue = []
    eft = 'EFT'
    
    for j in range(1,4):
        excited = None
        vividness = None
        if j == 1:
            delay = 30
            ip = IP30[i]
            if GROUP[i] == 1:
                cue = MONTH_G1_CUE[i]
                excited = exciting_G1_month[i]
                vividness = vivid_G1_month[i]
            else:
                cue = MONTH_G2_CUE[i]
                excited = exciting_G2_month[i]
                vividness = vivid_G2_month[i]
        elif j == 2:
            delay = 180
            ip = IP180[i]
            if GROUP[i] == 1:
                cue = MONTH_G1_CUE[i]
                excited = exciting_G1_month[i]
                vividness = vivid_G1_month[i]
            else:
                cue = MONTH_G2_CUE[i]
                excited = exciting_G2_month[i]
                vividness = vivid_G2_month[i]
        elif j == 3:
            delay = 365
            ip = IP365[i]
            if GROUP[i] == 1:
                cue = YEAR_G1_CUE[i]
                excited = exciting_G1_yr[i]
                vividness = vivid_G1_yr[i]
            else:
                cue = YEAR_G2_CUE[i]
                excited = exciting_G2_yr[i]
                vividness = vivid_G2_yr[i]
            
        ID.append(SID[i])
        EFT_ERT.append(eft)
        DELAY.append(delay)
        IP_VALUES.append(ip)
        IP.append('IP' + str(delay))
        CUES.append(cue)
        EXCITEMENT.append(excited)
        VIVID.append(vividness)
```

```python
compiled_data_study6 = pd.DataFrame({ 'ID' : ID,
                            'CATEGORY': EFT_ERT,
                            'DELAY': DELAY,
                            'IP': IP,
                            'IP VALUE': IP_VALUES,
                            'CUES': CUES,
                            'EXCITEMENT': EXCITEMENT,
                            'VIVIDNESS': VIVID })
compiled_data_study6.to_excel('cleaned/CLEAN_STUDY_6.xlsx', index = False)
```
